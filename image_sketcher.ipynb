```python
#to read the image
image= cv2.imread('68f093dd88a7753a738075954abfb101.jpg')

#to display the image
cv2_imshow(image)
```

```python
#converting the image to grayscale
grey_image= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

```python
#converting image to inverted grayscale

invert= cv2.bitwise_not(grey_image)
cv2_imshow(invert)
```

```python
blur_img=cv2.GaussianBlur(invert,(101,101),0)
cv2_imshow(blur_img)
```

```python
#inverting the blurred image

inv_blur=cv2.bitwise_not(blur_img)
cv2_imshow(inv_blur)
```

```python
sketch= cv2.divide(grey_image,inv_blur,scale=255.0)
cv2_imshow(sketch)
```

```python
cv2.imwrite("sketch.png" , sketch)
```
